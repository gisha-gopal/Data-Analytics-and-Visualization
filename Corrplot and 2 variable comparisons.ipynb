```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset
hi_df = pd.read_csv('DAV_Project_dataset.csv')
hi_df.head()
```

```python
hi_df.describe()
```

```python
#correlation matrix
hi_corr = hi_df.corr()
hi_corr
```

```python
import seaborn as sns
fig,ax = plt.subplots(figsize=(10,6))
sns.heatmap(hi_corr,annot=True,cmap="YlGnBu",ax=ax);
```

```python
%matplotlib inline
plt.style.use("seaborn-whitegrid")
fig, ax = plt.subplots(nrows=2,ncols=1,figsize = (20,15))
ax[0].scatter(hi_df["Happiness_Rank"], hi_df["Economy_GDP"],color="darkorchid");
ax[0].set(title = "Rank vs GDP per capita", ylabel = "GDP per capita")
ax[0].set_xlim(0,160);
ax[1].scatter(hi_df["Happiness_Rank"], hi_df["Family"], color="orange")
ax[1].set(title="Rank vs Social Support", xlabel="Rank",ylabel = "Social Suport");
ax[1].set_xlim(0,160);
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,figsize = (20,15))
ax[0].scatter(hi_df["Happiness_Rank"], hi_df["Generosity"],color="olive");
ax[0].set(title = "Rank vs Generosity", ylabel = "Generosity")
ax[0].set_xlim(0,160);
ax[1].scatter(hi_df["Happiness_Rank"], hi_df["Government_Corruption"], color="darkblue")
ax[1].set(title="Rank vs Corruption", xlabel="Rank",ylabel = "Government Corruption");
ax[1].set_xlim(0,160);
```
